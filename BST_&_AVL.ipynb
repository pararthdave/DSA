# Creating a Node

```python
class Node: 
    def __init__(self, key): 
        self.left = None
        self.right = None
        self.val = key 
        self.parent=None
```

# BINARY SEARCH TREE functions

1.Query operations:
search
minimum
maximum
successor
predecessor

2.modify operations:
insert
delete
left rotate
right rotate

3.traversal:
inorder(recursive)
preorder(recursive)
postorder(recursive)

4.node wise parameter:
depth 
height
no of nodes
rank

```python
class BST: 
    #SEARCH FUNCTION
    def search(self,root,key):
        if root.val==key or root is None:
            return root
        elif root.val<key:
            return self.search(root.right,key)
        else:
            return self.search(root.left,key)  
    #MINIMUM FUNCTION
    def minimum(self,x):
        while x.left is not None:
            x=x.left
        return x
    #MAXIMUM FUNCTION
    def maximum(self,x):
        while x.right is not None:
            x=x.right
        return x
    #SUCCESSOR FUNCTION
    def successor(self,x):
        if x.right is not None:
            return self.minimum(x.right)
        while x.parent is not None and x==x.parent.right:
                x=x.parent
        return x.parent
    #PREDECESSOR
    def predecessor(self,x):
        if x.left is not None:
            return self.maximum(x.left)
        while x.parent is not None and x==x.parent.left:
                x=x.parent
        return x.parent
    #modify operations
    #INSERT FUNCTION
    def insert(self,root, key): 
        if root is None: 
            return Node(key) 
        else: 
            if root.val == key:
                print(root.val)
                return root 
            elif root.val < key:
                root.right = self.insert(root.right, key)
                root.right.parent=root
            else:
                root.left = self.insert(root.left,key) 
                root.left.parent=root
        return root
    #DELETE FUNCTION
    def delete(self,root,key):
        x=self.search(root,key)
        if x.left is None and x.right is None:
            if x.parent is None:
                root=None
                return root
            else:
                if x.parent.left is not None and x.parent.left.val==x.val:
                    x.parent.left=None
                    x=x.parent
                else:
                    x.parent.right=None
                    x=x.parent
            while x.parent is not None:
                x=x.parent
            return x
        if x.left!=None and x.right==None:
            if x.parent is None:
                root=x.left
                x=None
                return root
            else:
                x.left.parent=x.parent
                x.parent.left=x.left
                while x.parent is not None:
                    x=x.parent
                return x     
        y=self.successor(x)
        if y.val==x.right.val:
            y.left=x.left
            if x.left is not None:
                x.left.parent=y
            y.parent=x.parent
            if y.parent is None:
                root=y
            else:
                if x==x.parent.left:
                    x.parent.left=y
                else:
                    x.parent.right=y
            while y.parent is not None:
                y=y.parent
            return y
        else:
            y.parent.left=y.right
            y.right.parent=y.parent
            y.parent=x.parent
            y.left=x.left
            y.right=x.right
            while y.parent is not None:
                y=y.parent
            return y
    #LEFT ROTATE
    def left_rotate(self,root,x):
        y=x.right
        z=x.parent
        x.right=y.left
        if y.left is not None:
            y.left.parent=x
        y.parent=z
        y.left=x
        x.parent=y
        if z is None:
            root=y
        else:
            if x==z.left:
                z.left=y
            else:
                z.right=y
        return y
    #RIGHT ROTATE
    def right_rotate(self,root,x):
        y=x.left
        z=x.parent
        x.left=y.right
        if y.right is not None:
            y.right.parent=x
        y.parent=z
        y.right=x
        x.parent=y
        if z is None:
            root=y
        else:
            if x==z.left:
                z.left=y
            else:
                z.right=y
        return y
    #INORDER TRAVERSAL
    def inorder(self,root): 
        if root: 
            self.inorder(root.left) 
            print(root.val)
            self.inorder(root.right) 
    #PREORDER TRAVERSAL
    def preorder(self,root): 
        if root: 
            print(root.val)
            self.preorder(root.left) 
            self.preorder(root.right)
    #POSTORDER TRAVERSAL
    def postorder(self,root): 
        if root: 
            self.postorder(root.left) 
            self.postorder(root.right)
            print(root.val)
    #DEPTH
    def depth(self,x):
        if x.parent is None:
            return 0
        dep=self.depth(x.parent)+1
        print(x.val)
        return dep
    #HEIGHT
    def height(self,x):
        if x is None:
            return -1
        if x.left is None and x.right is None:
            return 0
        h=max(self.height(x.left),self.height(x.right))+1
        return h
    #NO_OF_NODES
    def no_of_nodes(self,x):
            if x is None:
                return 0
            if x.left is None and x.right is None:
                return 1
            else:
                return self.no_of_nodes(x.left)+self.no_of_nodes(x.right)+1
    #GLOBAL RANK
    def rank(self,root,x):
        y=self.minimum(root)
        rank=1
        if y.val==x:
            return rank
        while y is not None and y.val!=x:
            y=self.successor(y)
            rank+=1
        return rank
```

# AVL TREES

```python
class AVL(BST):
    def balance(self,x):
        return self.height(x.right)-self.height(x.left)
    def insert_restore(self,root,x):
        val=x
        x=self.insert(root,x)
        x=self.search(x,val)
        while ((x.parent is not None) and (self.balance(x)==0 or
                                           self.balance(x)==1 or 
                                           self.balance(x)==-1)):
            x=x.parent
        if x is not None:
            balx=self.balance(x)
            if balx==-2:
                if self.balance(x.left)==1:
                    x=self.left_rotate(root,x.left)
                    x=x.parent
                x=self.right_rotate(root,x)
            elif balx==2:
                if self.balance(x.right)==-1:
                    x=self.right_rotate(root,x.right)
                    x=x.parent
                x=self.left_rotate(root,x)  
        while x.parent is not None:
             x=x.parent
        return x
    def is_avl(self,root):
        if root: 
            x=self.balance(root)
            if x==0 or x==1 or x==-1:
                self.is_avl(root.left)
                self.is_avl(root.right)
            else:
                return False
        return True
```
